# A ) Edge based segmentation

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import data, color
from skimage.filters import sobel
from skimage.feature import canny
from scipy import ndimage as ndi

# Load the four images
image1 = color.rgb2gray(plt.imread("photo1.jpg"))
image2 = color.rgb2gray(plt.imread("photo2.jpg"))
image3 = color.rgb2gray(plt.imread("photo3.jpg"))
image4 = color.rgb2gray(plt.imread("photo4.jpg"))

# Display the original images
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))
ax = axes.ravel()
for i in range(4):
    ax[i].imshow(eval(f"image{i+1}"), cmap=plt.cm.gray)
    ax[i].set_axis_off()
plt.tight_layout()

# Perform edge-based segmentation on the images using Canny detector
edges1 = canny(image1, sigma=3, low_threshold=0.1, high_threshold=0.2)
fill1 = ndi.binary_fill_holes(edges1)
edges2 = canny(image2, sigma=3, low_threshold=0.1, high_threshold=0.2)
fill2 = ndi.binary_fill_holes(edges2)
edges3 = canny(image3, sigma=3, low_threshold=0.1, high_threshold=0.2)
fill3 = ndi.binary_fill_holes(edges3)
edges4 = canny(image4, sigma=3, low_threshold=0.1, high_threshold=0.2)
fill4 = ndi.binary_fill_holes(edges4)

# Display the results
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))
ax = axes.ravel()
for i in range(4):
    ax[i].imshow(eval(f"fill{i+1}"), cmap=plt.cm.gray)
    ax[i].set_axis_off()
    ax[i].set_title(f"Filled edges for image {i+1}")
plt.tight_layout()

```
