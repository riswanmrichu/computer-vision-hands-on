# Spot the difference between two images

```python
import matplotlib.pyplot as plt
from PIL import Image, ImageChops

# Open the two input images
img1 = Image.open("spot 1.jpg")
img2 = Image.open("spot 2.jpg")

# Show the two input images using Matplotlib
plt.subplot(121)
plt.imshow(img1)
plt.title('Image 1')

plt.subplot(122)
plt.imshow(img2)
plt.title('Image 2')

plt.show()

# Calculate the difference between the two images
diff = ImageChops.difference(img1, img2)

# If there is any difference, show the difference image using Matplotlib
if diff.getbbox():
    plt.imshow(diff)
    plt.title('Difference')
    plt.show()
```
