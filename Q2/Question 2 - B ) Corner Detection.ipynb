# B ) Corner Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def detect_corners(image_path):
    # load the input image
    img = cv2.imread(image_path)

    # convert the input image into grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # modify the data type setting to 32-bit floating point
    gray = np.float32(gray)

    # apply the cv2.cornerHarris method to detect the corners
    corners = cv2.cornerHarris(gray, 2, 3, 0.05)

    #result is dilated for marking the corners
    corners = cv2.dilate(corners, None)

    # Threshold for an optimal value.
    img[corners > 0.01 * corners.max()]=[0, 0, 255]

    # display the image with corners detected using Matplotlib
    fig, ax = plt.subplots(1, 1, figsize=(10, 5))
    ax.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    ax.scatter(np.where(corners > 0.01 * corners.max())[1], np.where(corners > 0.01 * corners.max())[0], s=5, c='r')
    plt.show()

# call the function for each image
image_paths = ["photo1.jpg", "photo2.jpg", "photo3.jpg", "photo4.jpg"]
for path in image_paths:
    detect_corners(path)
```
