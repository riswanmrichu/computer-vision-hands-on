#  3 ) SIFT & ORB

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
img = cv2.imread('photo3.jpg')

# Create transformation matrices for scaling, rotation, affine, and perspective transformations
scaling_matrix = np.float32([[1.5, 0, 0], [0, 1.5, 0], [0, 0, 1]])
rotation_matrix = cv2.getRotationMatrix2D((img.shape[1]/2, img.shape[0]/2), 30, 1)
affine_matrix = np.float32([[1, 0, 100], [0, 1, 50]]).reshape(2, 3)
perspective_matrix = np.float32([[1, 0.5, 0], [-0.2, 1, 0], [0, 0, 1]])

# Apply the transformation matrices to the original image to create four new images
scaled_img = cv2.warpPerspective(img, scaling_matrix, (img.shape[1], img.shape[0]))
rotated_img = cv2.warpAffine(img, rotation_matrix, (img.shape[1], img.shape[0]))
affine_matrix = affine_matrix.astype(np.float32) # specify data type as np.float32
affine_img = cv2.warpAffine(img, affine_matrix, (img.shape[1], img.shape[0]))
perspective_img = cv2.warpPerspective(img, perspective_matrix, (img.shape[1], img.shape[0]))

# Initialize SIFT and ORB detectors
sift = cv2.SIFT_create()
orb = cv2.ORB_create()

# Detect keypoints and compute descriptors for the original and transformed images using SIFT
kp1, des1 = sift.detectAndCompute(img, None)
kp2, des2 = sift.detectAndCompute(scaled_img, None)
kp3, des3 = sift.detectAndCompute(rotated_img, None)
kp4, des4 = sift.detectAndCompute(affine_img, None)
kp5, des5 = sift.detectAndCompute(perspective_img, None)

# Draw the keypoints on the images and display them using matplotlib
img1 = cv2.drawKeypoints(img, kp1, None)
img2 = cv2.drawKeypoints(scaled_img, kp2, None)
img3 = cv2.drawKeypoints(rotated_img, kp3, None)
img4 = cv2.drawKeypoints(affine_img, kp4, None)
img5 = cv2.drawKeypoints(perspective_img, kp5, None)

plt.subplot(2,3,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('Original')
plt.subplot(2,3,2),plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)),plt.title('SIFT - Original')
plt.subplot(2,3,3),plt.imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)),plt.title('SIFT - Scaled Image')
plt.subplot(2,3,4),plt.imshow(cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)),plt.title('SIFT - Rotated Image')
plt.subplot(2,3,5),plt.imshow(cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)),plt.title('SIFT - Affine Image')
plt.subplot(2,3,6),plt.imshow(cv2.cvtColor(img5,cv2.COLOR_BGR2RGB)),plt.title('SIFT - Perspective Image')
plt.show()

#Detect keypoints and compute descriptors for the original and transformed images using ORB
kp1, des1 = orb.detectAndCompute(img, None)
kp2, des2 = orb.detectAndCompute(scaled_img, None)
kp3, des3 = orb.detectAndCompute(rotated_img, None)
kp4, des4 = orb.detectAndCompute(affine_img, None)
kp5, des5 = orb.detectAndCompute(perspective_img, None)

#Draw the keypoints on the images and display them using matplotlib
img1 = cv2.drawKeypoints(img, kp1, None)
img2 = cv2.drawKeypoints(scaled_img, kp2, None)
img3 = cv2.drawKeypoints(rotated_img, kp3, None)
img4 = cv2.drawKeypoints(affine_img, kp4, None)
img5 = cv2.drawKeypoints(perspective_img, kp5, None)

plt.subplot(2,3,1),plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB)),plt.title('Original')
plt.subplot(2,3,2),plt.imshow(cv2.cvtColor(img1,cv2.COLOR_BGR2RGB)),plt.title('ORB - Original')
plt.subplot(2,3,3),plt.imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)),plt.title('ORB - Scaled Image')
plt.subplot(2,3,4),plt.imshow(cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)),plt.title('ORB - Rotated Image')
plt.subplot(2,3,5),plt.imshow(cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)),plt.title('ORB - Affine Image')
plt.subplot(2,3,6),plt.imshow(cv2.cvtColor(img5,cv2.COLOR_BGR2RGB)),plt.title('ORB - Perspective Image')
plt.show()
```
