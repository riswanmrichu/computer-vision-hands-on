# A ) Edge Detection

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def detect_edges(image):
    # convert it to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    # perform the canny edge detector to detect image edges
    edges = cv2.Canny(gray, threshold1=30, threshold2=100)

    return edges

# read the images
image_names = ["photo1.jpg", "photo2.jpg", "photo3.jpg", "photo4.jpg"]
images = [cv2.imread(name) for name in image_names]

# detect edges for each image
edge_images = [detect_edges(img) for img in images]

# display the original and edge-detected images side by side for each image
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(8, 12))

for i, (img, edge_img) in enumerate(zip(images, edge_images)):
    axes[i, 0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    axes[i, 0].set_xticks([])
    axes[i, 0].set_yticks([])
    axes[i, 1].imshow(edge_img, cmap="gray")
    axes[i, 1].set_xticks([])
    axes[i, 1].set_yticks([])
    
plt.tight_layout()
plt.show()
```
