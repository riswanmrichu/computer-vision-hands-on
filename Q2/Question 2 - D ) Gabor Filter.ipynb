# D ) Gabor Filter

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_gabor_filter(img, theta, lamda):
    # Define the kernel size and standard deviation of the Gaussian function
    ksize = 31
    sigma = 5
    
    # Create the Gabor kernel
    kernel = cv2.getGaborKernel((ksize, ksize), sigma, theta, lamda, 1, 0, ktype=cv2.CV_32F)
    
    # Apply the Gabor filter to the input image
    filtered = cv2.filter2D(img, cv2.CV_8UC3, kernel)
    
    return filtered

# Read the four input images
img1 = cv2.imread('photo1.jpg')
img2 = cv2.imread('photo2.jpg')
img3 = cv2.imread('photo3.jpg')
img4 = cv2.imread('photo4.jpg')

# Convert the input images to grayscale
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
gray3 = cv2.cvtColor(img3, cv2.COLOR_BGR2GRAY)
gray4 = cv2.cvtColor(img4, cv2.COLOR_BGR2GRAY)

# Apply Gabor filters to the input images with the specified parameters
filtered1 = apply_gabor_filter(gray1, 0, 20)
filtered2 = apply_gabor_filter(gray2, 0, 80)
filtered3 = apply_gabor_filter(gray3, np.pi/2, 30)
filtered4 = apply_gabor_filter(gray4, 5*np.pi/6, 10)

# Display the input and output images using Matplotlib
fig, ax = plt.subplots(2, 4, figsize=(12, 6))
ax[0, 0].imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
ax[0, 1].imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
ax[0, 2].imshow(cv2.cvtColor(img3, cv2.COLOR_BGR2RGB))
ax[0, 3].imshow(cv2.cvtColor(img4, cv2.COLOR_BGR2RGB))
ax[1, 0].imshow(filtered1, cmap='gray')
ax[1, 1].imshow(filtered2, cmap='gray')
ax[1, 2].imshow(filtered3, cmap='gray')
ax[1, 3].imshow(filtered4, cmap='gray')
ax[0, 0].set_title('Input Image 1')
ax[0, 1].set_title('Input Image 2')
ax[0, 2].set_title('Input Image 3')
ax[0, 3].set_title('Input Image 4')
ax[1, 0].set_title('Gabor Filtered (0 deg, 20 wl)')
ax[1, 1].set_title('Gabor Filtered (0 deg, 80 wl)')
ax[1, 2].set_title('Gabor Filtered (90 deg, 30 wl)')
ax[1, 3].set_title('Gabor Filtered (150 deg, 10 wl)')
plt.show()
```
